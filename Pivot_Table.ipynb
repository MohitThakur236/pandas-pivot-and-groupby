```python
import pandas as pd

# creating a small dataset
df = pd.DataFrame({
    "city": ["Delhi", "Delhi", "Mumbai", "Mumbai", "Delhi"],
    "category": ["Electronics", "Clothing", "Electronics", "Clothing", "Electronics"],
    "amount": [1000, 500, 700, 300, 2000]
})



#### pivot table basics


# total sales by city and category
df.pivot_table(
    values="amount",
    index="city",
    columns="category",
    aggfunc="sum",
    fill_value=0,     # replacing NaN with 0
    margins=True      # adds total row and column
)
# looks like Delhi has higher electronics sales compared to Mumbai


# multiple aggregations
df.pivot_table(
    values="amount",
    index="city",
    columns="category",
    aggfunc=["sum", "mean"],        # gives both total and average at the same time
)


# average sales
df.pivot_table(
    values="amount",
    index="city",
    columns="category",
    aggfunc="mean",
    fill_value=0,
    margins=True
)
# helps to understand average spending pattern



#### groupby vs pivot


# similar result using groupby (but in multi-index form)
df.groupby(["city", "category"])["amount"].sum()

# pivot table gives a cleaner tabular view compared to groupby


# converting groupby result into pivot-like format
df.groupby(["city", "category"])["amount"].sum().unstack()

# unstack spreads category into columns (like pivot table)



#### working with multi-index


# accessing specific value
df.groupby(["city", "category"])["amount"].sum().loc[("Delhi", "Electronics")]

# accessing all values of Delhi
df.groupby(["city", "category"])["amount"].sum().loc["Delhi"]



#### stack and unstack


# converting to table and back
df.groupby(["city", "category"])["amount"].sum().unstack().stack()

# unstack -> spreads data into columns
# stack -> compresses back to multi-index


# converting multi-index into normal dataframe
df.groupby(["city", "category"])["amount"].sum().reset_index()



#### handling NaN in stack


# by default stack drops NaN values
# to keep them:
df.stack(dropna=False)



#### level-based unstack


# here level 0 = city, level 1 = category
df.unstack(level=0)

# this changes the structure based on level
```
